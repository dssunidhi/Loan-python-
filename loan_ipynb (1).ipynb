```python
#Data manipulation
import pandas as pd
#Numercial Computation
import numpy as np
#Data Visulazation
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Load Data
loan_data=pd.read_csv("Loan.csv")
#check first 5 records
loan_data.head()
```

```python
#Check last 5 records
loan_data.tail()
```

```python
#check No.of rows and columns
loan_data.shape
```

```python
#check Data Type
loan_data.info()
```

```python
#check count,mean,min,max,std,qurtile values of numerical dataset
loan_data.describe()
```

```python
#check null values in dataset
loan_data.isnull().sum()
```

```python
#handle missing value categorical varibles
loan_data['Gender'].fillna(loan_data['Gender'].mode()[0],inplace=True)
loan_data['Married'].fillna(loan_data['Married'].mode()[0],inplace=True)
loan_data['Dependents'].fillna(loan_data['Dependents'].mode()[0],inplace=True)
loan_data['Self_Employed'].fillna(loan_data['Self_Employed'].mode()[0],inplace=True)
#handle missing value with numerical varibles
loan_data['LoanAmount'].fillna(loan_data['LoanAmount'].mean(),inplace=True)
loan_data['Loan_Amount_Term'].fillna(loan_data['Loan_Amount_Term'].mean(),inplace=True)
loan_data['Credit_History'].fillna(loan_data['Credit_History'].mean(),inplace=True)
```

```python
loan_data.isnull().sum()
```

```python
loan_data['Total Income']=loan_data['ApplicantIncome']+loan_data['CoapplicantIncome']
loan_data.head()
```

```python
#drop unnecessary column
loan_data=loan_data.drop(columns=['Loan_ID'])
```

```python
#check reltionship between numerical varibles
sns.pairplot(loan_data)
```

```python
# Distribution of categorical variables
categorical_vars = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']
```

```python
# Plotting histograms for numerical variables
numerical_vars = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']

```

```python
# Analyzing Loan Status
sns.countplot(x='Loan_Status', data=loan_data)
plt.show()

# Analyzing the relationship between Loan Status and other variables
for var in categorical_vars:
    if var != 'Loan_Status':
        sns.countplot(x=var, hue='Loan_Status', data=loan_data)
        plt.show()

# Numerical variables vs Loan Status
for var in numerical_vars:
    sns.boxplot(x='Loan_Status', y=var, data=loan_data)
    plt.show()
```
